```python
import gradio as gr
from paraphraser import generate_paraphrase
from entailment import analyze_entailment
from highlighter import highlight_common_words_dict
from lcs import find_common_subsequences
from tree import generate_plot

def process_text(text, threshold):
    paraphrased_sentences = generate_paraphrase(text)
    all_sentences, selected_sentences, discarded_sentences = analyze_entailment(text, paraphrased_sentences, threshold)
    common_words = find_common_subsequences(text, list(selected_sentences.keys()))
    highlighted_text = highlight_common_words_dict(common_words, selected_sentences, discarded_sentences, "Paraphrased Sentences")

    highlighted_user_prompt = highlight_common_words_dict(common_words, {text: None}, {}, "User Prompt")


    selected_sentences_list = list(selected_sentences.keys())[:5]
    plots = [generate_plot(text, [sentence]) for sentence in selected_sentences_list]

    main_plot = generate_plot(text, selected_sentences_list)

    return (highlighted_user_prompt, highlighted_text, main_plot) + tuple(selected_sentences_list) + tuple(plots)

with gr.Blocks(theme=gr.themes.Monochrome()) as demo:
    gr.Markdown("# AIISC Watermarking Model")

    with gr.Row():
        input_text = gr.Textbox(label="User Prompt", lines=3)

    with gr.Row():
        submit_btn = gr.Button("Submit")
        clear_btn = gr.Button("Clear")

    threshold = gr.Slider(minimum=0, maximum=1, value=0.7, label="Entailment Threshold", visible=False)

    with gr.Row():
        highlighted_user_prompt = gr.HTML(label="User Prompt (Highlighted)")

    with gr.Row():
        output_text = gr.HTML(label="Analysis Result")

    with gr.Row():
        main_plot = gr.Plot(label="Main Plot")

    with gr.Tabs() as tabs:
        sentence_textboxes = []
        plot_outputs = []
        for i in range(5):
            with gr.Tab(f"Sentence {i+1}"):
                sentence_textboxes.append(gr.Textbox(label=f"Selected Sentence {i+1}", interactive=False))
                plot_outputs.append(gr.Plot(label=f"Plot for Sentence {i+1}"))

    submit_btn.click(
        process_text,
        inputs=[input_text, threshold],
        outputs=[highlighted_user_prompt, output_text, main_plot] + sentence_textboxes + plot_outputs
    )

    def clear_outputs():
        return [""] * 2 + [None] * 7  # Clear input, output_text, main_plot, 5 sentence textboxes, and 5 plots

    clear_btn.click(clear_outputs, inputs=None, outputs=[input_text, highlighted_user_prompt, output_text, main_plot] + sentence_textboxes + plot_outputs)

demo.launch()
```
